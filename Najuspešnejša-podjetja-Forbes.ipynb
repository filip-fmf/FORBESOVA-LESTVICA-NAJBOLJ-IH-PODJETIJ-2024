### Kako se podjetja med seboj razlikujejo glede na dobiček?

<p style='text-align: justify;'>
Podatki v datoteki <i>podatki_2025.csv</i> so zelo obsežni, zato sem se osredotočil le na naključna podjetja. Nakjlučno sem izbral podjetja in za te ustvaril tabelo in graf, ki prikazuje dobiček različnih podjetij. Poljubno sem izbral indekse podjetij, in sicer v naraščajočem vrstnem redu. 
</p>

```python
df = pd.read_csv("podatki/podatki_2025.csv")
df.index = df.index + 1

izbrani_stolpci = ["Ime podjetja", "Dobiček"]
izbrane_vrstice = [10, 87, 266, 383, 1304, 1467, 2000]          
                # če želimo drugo analizo, enostavno spremenimo številke, ob zagonu bomo dobili novo tabelo in diagram z različnimi državami
tabela = df.loc[izbrane_vrstice, izbrani_stolpci]
print(tabela)

#graf
plt.figure(figsize=(12, 5))
plt.bar(tabela['Ime podjetja'], tabela['Dobiček'], color='skyblue')
plt.title("Dobiček izbranih podjetij (v miljonih USD $)")
plt.xlabel("Podjetje")
plt.ylabel("Dobiček (miljonih USD $)")
plt.xticks(rotation=45, ha='right')  # da se imena podjetij ne prekrivajo
plt.tight_layout()  # da graf ni stisnjen
plt.show()
```

<p style='text-align: justify;'>Iz histograma lahko vidimo, da podjetja z najmanjšim indeksom imajo največji dobiček, ta so na začetku x osi. Bolj kot se indeks veča, (premikamo se v desno po x-osi) bolj dobički podjetij padajo.
</p>


### Dobiček podjetij - grafični prikaz 2
Spodnji program naključno izbere 20 podjetij iz datoteke podatki_2025.csv in naredi stolpični diagram po padajočem dobičku. Vsakič ko zaženemo program, dobimo različen histogram.

```python
df = pd.read_csv("podatki/podatki_2025.csv")

# naključno izbere 20 podjetij (size= ...vpišemo koliko podjetij želimo...)
random_companies = np.random.choice(df['Ime podjetja'], size=20, replace=False)
tabela = df[df['Ime podjetja'].isin(random_companies)]

# organizira podjetja po padajočem dobičku
tabela = tabela.sort_values(by='Dobiček', ascending=False)

plt.figure(figsize=(10, 5))
plt.bar(tabela['Ime podjetja'], tabela['Dobiček'])
plt.title("Dobiček naključnih podjetij")
plt.xlabel("Podjetje")
plt.ylabel("Dobiček")
plt.xticks(rotation='vertical')
plt.show()
```

Ta stolpični diagram najbolje prikaže dobiček. Program ob vsakem zagonu ustvari nov diagram, iz katerega je najboljše vidna razlika med dobički različnih podjetij, saj je lažje berljiv in se izpišejo tudi imena podjetij.

### S katero industrijo se ukvarjajo podjetja?
<p style='text-align: justify;'>
Spodaj so prikazani podatki iz datoteke <i>podjetja_podrobnosti.csv</i>. 
Spodnji program je analiziral industrijo, s katero se ukvarjajo podjetja. Ker je spletna stran v angleščini so tudi podatki analizirani v angleščini. Program naredi tortni diagram, ki predstavlja vsa področja industrije, kot so na primer medicinski pripomočki, biotehnologija, pogonska sredstva, elektrotehnika in druge. Ob strani grafikona so napisana vsa področja, znotraj tortnega diagrama pa delež industrije med vsemi podjetji.
</p>

```python
df = pd.read_csv("podatki/podjetja_podrobnosti.csv")

industry_counts = df['Industry'].value_counts()
industry_labels = industry_counts.index
industry_sizes = industry_counts.values

plt.figure(figsize=(8, 8))
plt.pie(industry_sizes, labels=None, autopct='%1.1f%%', textprops={'fontsize': 7})
plt.title("Deleži industrije")

# ustvari legendo zunaj grafa
plt.legend(industry_labels, loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

<p style='text-align: justify;'>
Ob strani so napisana vsa področja industrije. Sledijo si padajočem vrstnem redu glede na število podjetij v tej industriji. Iz podatkov je razvidno, da je največji delež podjetij v zdravilski proizvodnji, internetnih vsebinah in informacijah, ter naftni in plinski proizvodnji. 
<br>

OPOMBA: ti podatki so le o prvih 50 podjetij na tem seznamu s spletne strani <a href="https://disfold.com/world/companies/">https://disfold.com/world/companies/</a>.
</p>

<p style='text-align: justify;'>
Spodnja tortna diagrama prikazujeta deleže industrije, s katerimi se ukvarja le tistih deset podjetij, ki so na začetku in na dnu seznama/tabele.
</p>

```python
df = pd.read_csv("podatki/podjetja_podrobnosti.csv")

# Prvi graf
industry_counts = df['Industry'].value_counts().nlargest(10)
industry_labels = industry_counts.index
industry_sizes = industry_counts.values

fig = plt.figure(figsize=(16, 12))

ax1 = fig.add_subplot(121)
ax1.pie(industry_sizes, labels=industry_labels, autopct='%1.1f%%')
ax1.set_title("Delež desetih podjetij na začetku seznama")

# Drugi graf
industry_counts = df['Industry'].value_counts().nsmallest(10)
industry_labels = industry_counts.index
industry_sizes = industry_counts.values

ax2 = fig.add_subplot(122)
wedges, texts, autotexts = ax2.pie(industry_sizes, labels=None, autopct='%1.1f%%', textprops={'fontsize': 8})
ax2.set_title("Delež desetih podjetij na dnu seznama")

# Puščice
for i, wedge in enumerate(wedges):
    angle = (wedge.theta2 - wedge.theta1) / 2.0 + wedge.theta1
    x = 0.75 * wedge.r * np.cos(np.deg2rad(angle))
    y = 0.75 * wedge.r * np.sin(np.deg2rad(angle))
    xytext = (1.2 * wedge.r * np.cos(np.deg2rad(angle)), 1.2 * wedge.r * np.sin(np.deg2rad(angle)))
    ax2.annotate(f"{industry_labels[i]}", (x, y), xytext=xytext, arrowprops=dict(arrowstyle="->"))

plt.tight_layout()
plt.show()
```

#### Na katerem področju/sektorju delujejo podjetja?
<p style='text-align: justify;'>
Spodaj je program, ki ustvari dva stolpična diagrama, ki sicer predstavljata isto stvar (iste podatke), le da je en barven, drugi pa enobarven. 
Graf prikazuje področja na katerem delujejo podjetja, kot so na primer tehnologija, energija, zdravstvo in druge.
</p>

```python
df = pd.read_csv("podatki/podjetja_podrobnosti.csv")
industry_counts = df['Sector'].value_counts()

# Prvi graf
plt.figure(figsize=(14, 5))
plt.subplot(121)
industry_counts.plot(kind='bar', color=['red', 'blue', 'green', 'orange', 'purple'])
plt.title("Deleži - področja podjetij")
plt.xlabel("Področje (sector)")
plt.ylabel("Število podjetij")
plt.xticks(rotation=90)

# Drugi graf
plt.subplot(122)
industry_counts.plot(kind='bar')
plt.title("Deleži - področja podjetij")
plt.xlabel("Področje (sector)")
plt.ylabel("Število podjetij")
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

Iz histograma vidimo, da se največ podjetij ukvarja na področju zdravja in tehnologije.

### Kakšna je skupna tržna vrednost vseh podjetij v posamezni državi?

Spodnji program analizira tržne vrednosti vseh podjetij v posameznih državah.

```python
df = pd.read_csv("podatki/podatki_2025.csv")

# vsota tržnih vrednosti po državah
vsote_trznih_vrednosti = df.groupby("Država")["Tržna vrednost"].sum().reset_index()

# povprečje tržnih vrednosti
povprecje = vsote_trznih_vrednosti["Tržna vrednost"].sum() / len(vsote_trznih_vrednosti)

# Razvrsti podatke padajoče
vsote_trznih_vrednosti = vsote_trznih_vrednosti.sort_values("Tržna vrednost", ascending=True) # če vpišemo False, se obrne

# vertikalni stolpični diagram
plt.figure(figsize=(12, 5))
plt.bar(vsote_trznih_vrednosti["Država"], vsote_trznih_vrednosti["Tržna vrednost"])
plt.xlabel('Država')
plt.ylabel('Vsota tržnih vrednosti')
plt.title('Vsota tržnih vrednosti po državah (2025)')

# Doda črto za povprečje
plt.axhline(povprecje, color='red', linestyle='--', label='Povprečje')

plt.legend()

# Obrni imena držav za 90 stopinj
plt.xticks(rotation=90)

plt.show()
```

<p style='text-align: justify;'>
Program je seštel vse tržne vrednosti vseh podjetij v posameznih državah. In jih je uredil v naraščajočem vrstnem redu. Nariše tudi povprečje vseh tržnih vrednosti, povprečje je narisano s črtkano rdečo črto. Iz grafa se hitro vidi, da je tržna vrednost podjetij največja v Združenih državah Amerike.
